```python
import dbconfig
from collections import defaultdict
from pathlib import Path
from sqlalchemy import create_engine, text  
```

```python
DB_URL = f"postgresql+psycopg2://{dbconfig.DB_USER}:{dbconfig.DB_PASS}@{dbconfig.DB_HOST}:{dbconfig.DB_PORT}/{dbconfig.DB_NAME}?sslmode=require"

engine = create_engine(DB_URL)
print("Connected to the database successfully.")
```

```python
def run(query):
    with engine.connect() as conn:
        result = conn.execute(text(query))
        return result.fetchall()
```

```python
with engine.begin() as conn:

    conn.execute(text("""
    CREATE TABLE IF NOT EXISTS f_supplement_tracker (
        id SERIAL PRIMARY KEY,
        date DATE,
        section_of_day TEXT,
        supplement_name TEXT,
        dose INTEGER,
        uom TEXT,
        note TEXT
    )
    """))

    conn.execute(text("""
    CREATE TABLE IF NOT EXISTS d_supplement_info (
        id SERIAL PRIMARY KEY,
        supplement_name TEXT,
        category TEXT,
        usage_description TEXT,
        uom TEXT
    )
    """))

    conn.execute(text("""
    CREATE TABLE IF NOT EXISTS d_section_of_day (
        id SERIAL PRIMARY KEY,
        section_name TEXT,
        start_time TIME,
        end_time TIME,
        uom TEXT
    )
    """))
```

```python
all_schemas = run("""
SELECT *
FROM INFORMATION_SCHEMA.TABLES
WHERE TABLE_SCHEMA = 'public';
""")

(all_schemas)
```

```python
conn.close()
```
